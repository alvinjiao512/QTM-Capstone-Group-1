```python
import ee
import geemap.core as geemap
ee.Authenticate()
ee.Initialize(project='ee-alvinqwq512')
```

```python
dataset = (ee.ImageCollection('USDA/NAIP/DOQQ')
           .filter(ee.Filter.date('2021-01-01', '2022-12-31')))

# Select the true color bands
true_color = dataset.select(['R', 'G', 'B'])

# Visualization parameters
true_color_vis = {
    'min': 0,
    'max': 255,
}

# Set the map center
center = [-84.286574, 33.794887]

# Create a Map
Map = geemap.Map()
Map.setCenter(center[0], center[1], 18)

# Add the true color layer to the map
Map.addLayer(true_color, true_color_vis, 'True Color')

# Display the map
Map
```
